```python
import yaml

# Step 1: Define the training and validation paths for Kaggle environment
train_paths = [
    '/kaggle/input/grid-flip/sss/Oppo Enco Buds2/train/images',
    '/kaggle/input/grid-flip/sss/Nabati Wafer/train/images',
    '/kaggle/input/grid-flip/sss/Minimalist Sunscreen/train/images',
    '/kaggle/input/grid-flip/sss/Maggie/train/images',
    '/kaggle/input/grid-flip/sss/Johnsons Baby Cream/train/images',
    '/kaggle/input/grid-flip/sss/Himalaya Skin Cream/train/images',
    '/kaggle/input/grid-flip/sss/Dettol/train/images',
    '/kaggle/input/grid-flip/sss/Denver Deodorant/train/images',
    '/kaggle/input/grid-flip/sss/Dabur Honey/train/images',
    '/kaggle/input/grid-flip/sss/Dabur Chyawanprash/train/images',
    '/kaggle/input/grid-flip/sss/Citizen Notebook/train/images',
    '/kaggle/input/grid-flip/sss/biotique sunscreen/train/images',
    '/kaggle/input/grid-flip/sss/Apsara Pencil/train/images',
    '/kaggle/input/grid-flip/sss/Apis Ginger Garlic Paste/train/images',
    '/kaggle/input/grid-flip/sss/Amul Butter/train/images'
]

val_paths = [
    '/kaggle/input/grid-flip/sss/Oppo Enco Buds2/valid/images',
    '/kaggle/input/grid-flip/sss/Nabati Wafer/valid/images',
    '/kaggle/input/grid-flip/sss/Minimalist Sunscreen/valid/images',
    '/kaggle/input/grid-flip/sss/Maggie/valid/images',
    '/kaggle/input/grid-flip/sss/Johnsons Baby Cream/valid/images',
    '/kaggle/input/grid-flip/sss/Himalaya Skin Cream/valid/images',
    '/kaggle/input/grid-flip/sss/Dettol/valid/images',
    '/kaggle/input/grid-flip/sss/Denver Deodorant/valid/images',
    '/kaggle/input/grid-flip/sss/Dabur Honey/valid/images',
    '/kaggle/input/grid-flip/sss/Dabur Chyawanprash/valid/images',
    '/kaggle/input/grid-flip/sss/Citizen Notebook/valid/images',
    '/kaggle/input/grid-flip/sss/biotique sunscreen/valid/images',
    '/kaggle/input/grid-flip/sss/Apsara Pencil/valid/images',
    '/kaggle/input/grid-flip/sss/Apis Ginger Garlic Paste/valid/images',
    '/kaggle/input/grid-flip/sss/Amul Butter/valid/images'
]

# Step 2: Define the number of classes and class names
nc = 15
names = [
    'Oppo Enco Buds2',
    'Nabati Wafer',
    'Minimalist Sunscreen',
    'Maggie',
    'Johnsons Baby Cream',
    'Himalaya Skin Cream',
    'Dettol',
    'Denver Deodorant',
    'Dabur Honey',
    'Dabur Chyawanprash',
    'Citizen Notebook',
    'biotique sunscreen',
    'Apsara Pencil',
    'Apis Ginger Garlic Paste',
    'Amul Butter'
] 

# Step 3: Create the configuration dictionary
config = {
    'train': train_paths,
    'val': val_paths,
    'nc': nc,
    'names': names
}

# Step 4: Specify the file path where you want to save the YAML file in Kaggle
yaml_file_path = '/kaggle/working/yoo_config.yaml'

# Step 5: Write the configuration to the YAML file
with open(yaml_file_path, 'w') as yaml_file:
    yaml.dump(config, yaml_file, default_flow_style=False)

print(f'YAML configuration file created: {yaml_file_path}')
```

```python
import yaml
import torch
from ultralytics import YOLO

# Load YAML configuration
yaml_file_path = '/kaggle/working/yoo_config.yaml'
with open(yaml_file_path, 'r') as yaml_file:
    config = yaml.safe_load(yaml_file)

# Load the YOLO model
model = YOLO('yolov8n.pt')  # Load the model without specifying device

# Move the model to the GPU if available
if torch.cuda.is_available():
    model.model.to('cuda')

# Train the model
results = model.train(
    data=yaml_file_path,
    epochs=10,
    imgsz=640,
    batch=15,
    lr0=0.001,  # Adjust learning rate if needed
    workers=8
)

# Print class counts after training
print("Training complete!")

# Accessing and printing class counts for each epoch
for epoch_result in results:
    print("Epoch results:")
    print(f"Class counts: {epoch_result.metrics['class_counts']}")  # Adjust based on the actual structure of results
```
